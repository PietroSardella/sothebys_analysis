## Análise de Dados sobre Artistas Contemporâneos

### Introdução

Este projeto visa explorar e analisar os dados de artistas contemporâneos utilizando o conjunto de dados disponível do Tate, que contém informações sobre artistas e suas obras. O foco da análise é entender a distribuição de gêneros, a idade dos artistas e as nacionalidades dos mesmos, especialmente aqueles nascidos após 1950.

### Objetivos do Projeto

1. **Visualizar a Distribuição de Gêneros**: Analisar a proporção de artistas masculinos e femininos e sua evolução ao longo do tempo.
2. **Analisar a Idade dos Artistas**: Calcular a idade dos artistas contemporâneos e visualizar sua distribuição.
3. **Explorar Nacionalidade**: Identificar os países de origem mais comuns entre os artistas e analisar como o gênero está distribuído entre esses países.
4. **Criar Gráficos Informativos**: Utilizar gráficos para apresentar visualmente os dados e insights obtidos durante a análise.

### Dados Utilizados

Os dados utilizados neste projeto são:

- **artist_data.csv**: Contém informações sobre os artistas, incluindo nome, gênero, ano de nascimento e lugar de nascimento.

### Estrutura dos Dados
- `id` - id do artista            
- `name` - nome do artista         
- `gender` - gênero (Mulher, Homem)       
- `dates` - Ano de nascimento e falecimento        
- `yearOfBirth` - Ano de nascimento   
- `yearOfDeath` - Ano de falecimento  
- `placeOfBirth` - Local de nascimento 
- `placeOfDeath` - Local de falescimento 
- `url` - url do artista no site do Tate Museum     


### Metodologia

A análise será realizada em várias etapas, incluindo a leitura, limpeza e transformação dos dados, visualização dos dados e interpretação dos resultados. Utilizaremos bibliotecas como `pandas`, `seaborn` e `matplotlib` para facilitar a manipulação e visualização dos dados.

```python
# Verificar se há valores não numéricos (caso haja necessidade de conversão para números)
# Exibir linhas com valores não numéricos em 'yearOfBirth' e 'yearOfDeath'
print(artists[pd.to_numeric(artists['yearOfBirth'], errors='coerce').isna()])
print(artists[pd.to_numeric(artists['yearOfDeath'], errors='coerce').isna()])
```

```python
# Substituindo valores não numéricos por um valor específico, como uma média ou um valor nulo, por exemplo
artists['yearOfBirth'] = pd.to_numeric(artists['yearOfBirth'], errors='coerce')
artists['yearOfDeath'] = pd.to_numeric(artists['yearOfDeath'], errors='coerce')
```

```python
# Converter as colunas para o tipo float
artists['yearOfBirth'] = artists['yearOfBirth'].astype(float)
artists['yearOfDeath'] = artists['yearOfDeath'].astype(float)

# Preencher valores ausentes com a mediana
artists['yearOfBirth'] = artists['yearOfBirth'].fillna(artists['yearOfBirth'].median())
artists['yearOfDeath'] = artists['yearOfDeath'].fillna(artists['yearOfDeath'].median())

# Converter de volta para o tipo inteiro, se necessário
artists['yearOfBirth'] = artists['yearOfBirth'].astype(int)
artists['yearOfDeath'] = artists['yearOfDeath'].astype(int)
```

```python
# Verificar os tipos após a conversão
print(artists[['yearOfBirth', 'yearOfDeath']].dtypes)
```

```python
# Prencher valores nulos por 'Desconhecido'
artists.loc[artists['placeOfBirth'].isnull(), 'placeOfBirth'] = 'Unknown'
# Verificar se os valores nulos foram preenchidos
print(artists['placeOfBirth'].isnull().sum())  # Deve retornar 0

# Prencher valores nulos por 'Desconhecido'
artists.loc[artists['placeOfDeath'].isnull(), 'placeOfDeath'] = 'Unknown'
# Verificar se os valores nulos foram preenchidos
print(artists['placeOfDeath'].isnull().sum())  # Deve retornar 0

# Prencher valores nulos por 'Desconhecido'
artists.loc[artists['gender'].isnull(), 'gender'] = 'Unknown'
# Verificar se os valores nulos foram preenchidos
print(artists['gender'].isnull().sum())  # Deve retornar 0
```

#### Filtrar Dados por Período

```python
# Filtrando para artistas nascidos depois de 1950 e criando uma cópia
contemporary_artists = artists[artists['yearOfBirth'] >= 1950].copy()

# Preencher valores nulos em 'gender'
contemporary_artists['gender'] = contemporary_artists['gender'].fillna('Unknown')

# Contagem de artistas por gênero nascidos após 1950
gender_counts = contemporary_artists['gender'].value_counts()

# Criar um DataFrame para os dados de gênero contemporâneo
gender_contemporary = gender_counts.reset_index()
gender_contemporary.columns = ['gênero', 'número de artistas']
```

```python
# Configurar o estilo do gráfico
sns.set(style='whitegrid', rc={'axes.edgecolor': 'black'})

# Ordenar dados por número de artistas em ordem crescente
gender_contemporary = gender_contemporary.sort_values(by='número de artistas', ascending=False)

# Configurar figura
plt.figure(figsize=(10, 6))

# Criar uma paleta de cores azul com gradiente
blue_gradient = sns.color_palette("Blues", len(gender_contemporary))

# Plotar gráfico de barras vertical com cores em gradiente azul
bars = plt.bar(gender_contemporary['gênero'], gender_contemporary['número de artistas'], color=blue_gradient, edgecolor='black', alpha=0.85)

# Adicionar título e rótulos com estilo
plt.title('Número de Artistas por Gênero (Nascidos Após 1950)', fontsize=22, weight='bold', color='black', pad=20)
plt.ylabel('Número de Artistas', fontsize=16, color='black', labelpad=15)
plt.xlabel('Gênero', fontsize=16, color='black', labelpad=15)
plt.xticks(rotation=45, fontsize=14, color='black')  # Melhorar legibilidade
plt.yticks(fontsize=14, color='black')

# Adicionar anotações com o número de artistas acima de cada barra
for index, row in gender_contemporary.iterrows():
    plt.text(index, row['número de artistas'] + 2, str(row['número de artistas']),
             ha='center', color='black', fontsize=14, weight='bold', fontfamily='Arial')

# Estilizar o gráfico com uma grade leve
plt.grid(axis='y', linestyle='--', linewidth=0.8, color='gray', alpha=0.6)  # Grid mais suave
sns.despine(left=True, bottom=True)  # Remove o contorno esquerdo e inferior

# Ajuste do layout
plt.tight_layout()
plt.show()
```

#### Analisar Idade dos Artistas Contemporanêos - Nascidos apos 1950

```python
# Calcular idade dos artistas (considerando o ano atual se o ano de falecimento for NaN)
current_year = datetime.now().year
contemporary_artists['age'] = contemporary_artists.apply(
    lambda x: current_year - x['yearOfBirth'] if pd.isna(x['yearOfDeath']) else x['yearOfDeath'] - x['yearOfBirth'],
    axis=1
)
```

```python
# Visualizar distribuição etária
plt.figure(figsize=(10, 6))
sns.histplot(contemporary_artists['age'], bins=20, kde=True)
plt.title('Distribuição Etária dos Artistas Nascidos Após 1950')
plt.xlabel('Idade')
plt.ylabel('Número de Artistas')
plt.grid(True)
plt.show()
```

#### Explorar Gênero e Nacionalidade em Conjunto

```python
# Tabela cruzada de gênero e país de origem
gender_country = contemporary_artists.pivot_table(index='placeOfBirth', columns='gender', aggfunc='size', fill_value=0)

# Selecionar os 10 países com o maior número de artistas
country_counts = gender_country.sum(axis=1).nlargest(10)
top_countries = gender_country.loc[country_counts.index]

# Visualizar com gráficos de barras empilhadas
colors = ['#A7C6ED', '#40E0D0'] 
top_countries.plot(kind='barh', stacked=True, figsize=(10, 6), color=colors)  # Mudando para 'barh'
plt.title('Distribuição de Gênero e País de Origem (nascidos após 1950)')
plt.xlabel('Número de Artistas')  # Agora é o eixo X
plt.ylabel('País de Origem')  # Agora é o eixo Y
plt.legend(title='Gênero', labels=['Homens', 'Mulheres'])
plt.show()

```

```python
from urllib.error import HTTPError, URLError
from bs4 import BeautifulSoup
from urllib.request import urlopen

# URL da página da artista Fran Chang
url = "https://www.franchang.com/"

# Tentar abrir a página e tratar possíveis erros
try:
    page = urlopen(url)
except HTTPError as e:
    print(f"Erro HTTP: {e}")
except URLError as e:
    print(f"Erro URL: {e}")
else:
    # Parse do HTML
    soup = BeautifulSoup(page, "html.parser")
    
    # Listas para armazenar informações das obras
    titles = []
    descriptions = []
    dates = []
    images = []
    
    # Supondo que as obras estejam em tags específicas como divs ou seções com uma classe identificável
    # Vamos ajustar essa classe com base na estrutura do site da artista
    for artwork in soup.find_all("div", class_="artwork-class"):  # Substitua 'artwork-class' pela classe correta
        # Extrair o título da obra
        title = artwork.find("h2", class_="title-class").get_text(strip=True) if artwork.find("h2", class_="title-class") else None
        titles.append(title)
        
        # Extrair a descrição da obra
        description = artwork.find("p", class_="description-class").get_text(strip=True) if artwork.find("p", class_="description-class") else None
        descriptions.append(description)
        
        # Extrair a data da obra
        date = artwork.find("span", class_="date-class").get_text(strip=True) if artwork.find("span", class_="date-class") else None
        dates.append(date)
        
        # Extrair a URL da imagem, caso esteja em uma tag de imagem
        image = artwork.find("img", class_="image-class")["src"] if artwork.find("img", class_="image-class") else None
        images.append(image)
    
    # Criar o DataFrame com as informações coletadas
    artworks_df = pd.DataFrame({
        "Title": titles,
        "Description": descriptions,
        "Date": dates,
        "Image URL": images
    })
    
    # Exibir as primeiras linhas do DataFrame para verificar
    print(artworks_df.head())
```

```python
from bs4 import BeautifulSoup
import requests

# URL da página web com as obras de arte (exemplo fictício)
url = 'https://www.franchang.com/selected-works'

# Fazer a requisição à página
response = requests.get(url)
html_content = response.content

# Analisar o conteúdo HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Exemplo de estrutura para armazenar as informações das obras de arte
obras = []

# Encontrar cada obra de arte na página
for item in soup.find_all(class_='class-name-da-obra'):  # Substitua pelo nome correto da classe
    obra = {}
    
    # Extrair o título da obra
    titulo = item.find('h2', class_='titulo-obra')  # Substitua pela tag/classe correta
    obra['titulo'] = titulo.get_text(strip=True) if titulo else 'N/A'
    
    # Extrair a descrição da obra
    descricao = item.find('p', class_='descricao-obra')  # Substitua pela tag/classe correta
    obra['descricao'] = descricao.get_text(strip=True) if descricao else 'N/A'
    
    # Extrair a data de criação
    data_criacao = item.find('span', class_='data-criacao')  # Substitua pela tag/classe correta
    obra['data_criacao'] = data_criacao.get_text(strip=True) if data_criacao else 'N/A'
    
    # Extrair o preço
    preco = item.find('span', class_='preco-obra')  # Substitua pela tag/classe correta
    obra['preco'] = preco.get_text(strip=True) if preco else 'N/A'
    
    # Adicionar a obra à lista
    obras.append(obra)

# Exibir as informações extraídas
for obra in obras:
    print(f"Título: {obra['titulo']}")
    print(f"Descrição: {obra['descricao']}")
    print(f"Data de Criação: {obra['data_criacao']}")
    print(f"Preço: {obra['preco']}\n")
```
