# Análise de Dados sobre Artistas Contemporâneos

## Introdução

Este projeto visa explorar e analisar os dados de artistas contemporâneos utilizando o conjunto de dados disponível do Tate, que contém informações sobre artistas e suas obras. O foco da análise é entender a distribuição de gêneros, a idade dos artistas e as nacionalidades dos mesmos, especialmente aqueles nascidos após 1950.

## Objetivos do Projeto

1. **Visualizar a Distribuição de Técnicas e Suportes Utilizados**: Analisar a proporção de Técnicas e sua evolução ao longo do tempo.
2. **Analisar os maiores Artistas na Coleção**: Calcular artistas com maior quantidade de obra no Tate Museum e visualizar sua distribuição.
3. **Explorar Nacionalidade**: Identificar os países de origem mais comuns entre os artistas e analisar como o gênero está distribuído entre esses países.
4. **Criar Gráficos Informativos**: Utilizar gráficos para apresentar visualmente os dados e insights obtidos durante a análise.

## Dados Utilizados

Os dados utilizados neste projeto são:

- **artwork_data.csv**: Contém informações sobre as obras de arte, incluindo título, ano de criação, técnica utilizada e informações sobre a aquisição.

### Estrutura dos Dados

- **artwork_data.csv**:
    - `id`: Identificador único da obra
    - `artist`: Nome do artista
    - `title`: Título da obra
    - `year`: Ano em que a obra foi criada
    - `medium`: Técnica utilizada na obra

## Metodologia

A análise será realizada em várias etapas, incluindo a leitura dos dados, limpeza e transformação dos dados, visualização dos dados e interpretação dos resultados. Utilizaremos bibliotecas como `pandas`, `seaborn` e `matplotlib` para facilitar a manipulação e visualização dos dados.

Vamos começar!


```python
# Importando bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
```

```python
# Carregar o dataset
artwork = pd.read_csv(r'../artwork_data.csv', dtype={'year': str, 'height': str})
```

```python
# Visualizar os nomes das colunas
artwork.info()
```

```python
artwork['artist'].nunique()
```

```python
# Contar o número de obras por artista
artistas_contagem = artwork['artist'].value_counts()

# Selecionar os 10 artistas com mais obras
top_10_artistas = artistas_contagem.head(10)

# Criar um DataFrame para os dados
top_10_df = top_10_artistas.reset_index()
top_10_df.columns = ['artista', 'número de obras']
```

```python
# Configurar o estilo do gráfico com tema escuro
sns.set(style="white", font="Arial", font_scale=1.1)

# Criar uma paleta de cores
palette = sns.color_palette("mako", n_colors=10)

# Criar o gráfico de barras horizontal 
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=top_10_df, x='número de obras', y='artista', hue='artista', palette=palette, dodge=False, legend=False)

# Títulos e subtítulos
plt.title("Top 10 Artistas com Mais Obras", fontsize=18, weight='bold', pad=20)
plt.xlabel("Número de Obras", fontsize=14, labelpad=10)
plt.ylabel("Artista", fontsize=14, labelpad=10)

# Adicionar rótulos de dados ao final de cada barra
for index, value in enumerate(top_10_df['número de obras']):
    ax.text(value + 1, index, str(value), va='center', ha='left', color='black', fontsize=12)

# Remover bordas para um visual mais limpo
sns.despine(left=True, bottom=True)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

```python
print(artwork['acquisitionYear'].isnull().sum())  # Verifica o número de valores nulos

artwork = artwork[artwork['acquisitionYear'].notna()]  # Remove linhas com NaN na coluna

artwork['acquisitionYear'] = artwork['acquisitionYear'].astype(int)
```

```python
artwork.head()
```

```python
artwork_by_acquisitionYear = artwork['acquisitionYear'].value_counts().head(10)
artwork_by_acquisitionYear
```

```python
# Contar os artistas por técnicas utilizadas
artwork_by_medium = artwork['medium'].value_counts().head(10)
artwork_by_medium.head(10)
```

### 1. Técnicas mais utilizadas:

```python
# Configurar o estilo do gráfico
sns.set(style="white", font="Arial", font_scale=1.1)

# Obter as 10 técnicas mais utilizadas e ordenar em ordem decrescente
top_mediums = artwork['medium'].value_counts().head(10).sort_values(ascending=True)

# Definir uma paleta de cores gradiente
colors = sns.color_palette("coolwarm", n_colors=10)  # Paleta gradiente interessante

# Criar o gráfico de barras horizontal com gradiente
plt.figure(figsize=(10, 6))
bars = plt.barh(top_mediums.index, top_mediums.values, color=colors)

# Títulos e subtítulos com design aprimorado
plt.title('Top 10 Técnicas Utilizadas', fontsize=18, weight='bold', pad=20)
plt.xlabel('Número de Obras', fontsize=14, labelpad=10)
plt.ylabel('Técnica', fontsize=14, labelpad=10)

# Adicionar rótulos de dados ao final de cada barra
for index, value in enumerate(top_mediums):
    plt.text(value + 5, index, str(value), va='center', ha='left', color='black', fontsize=12)

# Aplicar um gradiente de cor às barras
for i, bar in enumerate(bars):
    bar.set_color(colors[i])

# Remover bordas para um visual mais limpo
sns.despine(left=True, bottom=True)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

```python
# Remover valores nulos na coluna 'acquisitionYear' para evitar problemas de contagem
artwork_clean = artwork.dropna(subset=['acquisitionYear'])

# Filtrar para o período a partir de 1820
acquisition_per_year = artwork_clean['acquisitionYear'].value_counts().sort_index()
acquisition_per_year = acquisition_per_year[acquisition_per_year.index >= 1820]

# Configurar estilo e cores
sns.set(style="whitegrid")
fig, ax = plt.subplots(figsize=(10, 6))

# Configurar gradiente de cores para a linha
norm = plt.Normalize(acquisition_per_year.index.min(), acquisition_per_year.index.max())
colors = plt.cm.viridis(norm(acquisition_per_year.index))

# Criar linhas e marcadores com gradiente
for i in range(len(acquisition_per_year) - 1):
    ax.plot(acquisition_per_year.index[i:i+2], acquisition_per_year.values[i:i+2], 
            color=colors[i], linewidth=2.5)

# Adicionar pontos marcadores e labels automáticos para picos
ax.scatter(acquisition_per_year.index, acquisition_per_year.values, color=colors, s=40, zorder=5)
for year, count in zip(acquisition_per_year.index, acquisition_per_year.values):
    if count > acquisition_per_year.mean() + acquisition_per_year.std():  # Exibir labels apenas em picos
        ax.text(year, count + 50, str(count), ha='center', fontsize=10, color='#4C4C4C')

# Ajustar título, rótulos e estilização
ax.set_title('Evolução Temporal da Aquisição de Obras - pós 1820', fontsize=18, weight='bold', pad=20, color='#4C4C4C')
ax.set_xlabel('Ano de Aquisição', fontsize=14, labelpad=10, color='#4C4C4C')
ax.set_ylabel('Número de Obras Adquiridas', fontsize=14, labelpad=10, color='#4C4C4C')
ax.set_xlim(1820, acquisition_per_year.index.max())  # Limitar a exibição do eixo x a partir de 1820
ax.tick_params(axis='x', rotation=45)
sns.despine(left=True, bottom=True)

# Estilo e espaçamento
ax.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Filtrar as obras do artista "Warhol, Andy"
warhol_artworks = artwork[artwork['artist'] == 'Warhol, Andy']

# Exibir as primeiras linhas para conferir
warhol_artworks.head()
```

```python
# Configurar o estilo do gráfico
sns.set(style='whitegrid')

# Configurar o tamanho da figura
plt.figure(figsize=(10, 6))

# Contar o número de obras por ano e ordenar
artworks_per_year = warhol_artworks['year'].value_counts().sort_index()

# Plotar o gráfico de linha com melhorias visuais
plt.plot(artworks_per_year.index, artworks_per_year.values, 
         marker='o', linestyle='-', color='royalblue', linewidth=2, markersize=8)

# Adicionar título e rótulos com estilo
plt.title('Obras de Andy Warhol ao longo dos anos', fontsize=22, weight='bold', color='black', pad=20)
plt.xlabel('Ano', fontsize=16, color='black', labelpad=10)
plt.ylabel('Número de Obras', fontsize=16, color='black', labelpad=10)

# Adicionar uma grade leve para melhor visualização
plt.grid(axis='y', linestyle='--', linewidth=0.7, color='gray', alpha=0.6)

# Adicionar anotações em cada ponto de dados
for x, y in zip(artworks_per_year.index, artworks_per_year.values):
    plt.text(x, y + 1, str(y), ha='center', fontsize=12, color='black', weight='bold')

# Melhorar os limites dos eixos
plt.xlim(artworks_per_year.index.min(), artworks_per_year.index.max())
plt.ylim(0, artworks_per_year.max() + 5)

# Definir ticks do eixo x para não sobrepor
plt.xticks(ticks=artworks_per_year.index[::2], fontsize=12, color='black')  # Exibir a cada 2 anos
plt.yticks(fontsize=12, color='black')

# Estilizar os eixos
plt.tick_params(axis='both', which='major', length=6, width=1, color='black')
plt.tick_params(axis='both', which='minor', length=4, width=1, color='gray')

# Ajustar o layout
plt.tight_layout()

# Mostrar o gráfico
plt.show()

```

```python
# Configurar o estilo do gráfico
sns.set(style='whitegrid')

# Configurar o tamanho da figura
plt.figure(figsize=(10, 6))

# Contar as principais técnicas usadas por Warhol e ordenar em ordem crescente
top_techniques_warhol = warhol_artworks['medium'].value_counts().head(10).sort_values(ascending=True)

# Definir uma paleta de cores gradiente
colors = plt.cm.Blues(range(0, 256, 25))  # Gradiente de azul

# Plotar o gráfico de barras horizontal com cores em gradiente
bars = plt.barh(top_techniques_warhol.index, top_techniques_warhol.values, color=colors)

# Adicionar bordas pretas
for bar in bars:
    bar.set_edgecolor('black')

# Título e rótulos ajustados
plt.title('Top 10 Técnicas Utilizadas nas Obras de Andy Warhol', fontsize=14, weight='bold')
plt.xlabel('Número de Obras', fontsize=12)
plt.ylabel('Técnica', fontsize=12)

# Adicionar anotações ao final de cada barra
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, 
             str(bar.get_width()), va='center', fontsize=10, color='black')

# Adicionar uma grade leve
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Ajustar o layout
plt.tight_layout()

# Exibir o gráfico
plt.show()

```
